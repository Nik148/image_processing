```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from skimage import io
```

```python
drive.mount('/content/drive')
url = "/content/drive/MyDrive/cats.jpg"
image = io.imread(url)
plt.imshow(image)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV_FULL)
filter_image = cv2.bitwise_not(hsv)


plt.figure(figsize=(18, 6))
filter_image = np.clip(filter_image * 1.3, 0, 255).astype(np.uint8)

filter_image = cv2.cvtColor(filter_image, cv2.COLOR_RGB2GRAY)


plt.subplot(1, 3, 1), plt.imshow(image), plt.title("Изображение")

plt.subplot(1, 3, 2), plt.imshow(filter_image, cmap="Blues"), plt.title("Измененное изображение")
```

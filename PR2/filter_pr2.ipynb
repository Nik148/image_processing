```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from skimage import io
```

```python
drive.mount('/content/drive')
url = "/content/drive/MyDrive/cats.jpg"
image = io.imread(url)
plt.imshow(image)
```

```python
filter = np.array([
  [-1, 1, 0, -3],
  [-2, 2, 0, -1],
  [-3, 1, 0, -1],
  [-1, 10, 0, -2]
])

filter_image = cv2.filter2D(image, -1, filter)

plt.figure(figsize=(18, 6))
plt.subplot(1, 2, 1), plt.imshow(image), plt.title("")
plt.subplot(1, 2, 2), plt.imshow(filter_image), plt.title(" ")
```
